```python
# Imports required
import os
from glob import glob
import matplotlib.pyplot as plt
import rasterio as rio
from rasterio.merge import merge
from rasterio.plot import show
```

```python
# Set working directory
os.chdir(os.path.join('X:/', '2020_fellowship', 'Georeferencing-mapping-aerial-images'))
```

```python
# Set paths to images
image_dir_path = glob(os.path.join('images', 'georeferenced', '*.tif'))
```

```python
# Let's first create an empty list for the datafiles that will be part of the mosaic.
src_files_to_mosaic = []

# Read each file with raterio and put in empty list above
for img in image_dir_path:
    src = rio.open(img)
    src_files_to_mosaic.append(src)

# Merge all the data into a single 'mosaic' array
mosaic, out_trans = merge(src_files_to_mosaic)

# Display the newly merger, mosaic image
fig, ax = plt.subplots(figsize=(5, 9))
show(mosaic, ax=ax)
```
